```python
import matplotlib.pyplot as plt
import numpy as np
from datasets import load_dataset
```

```python
cifar_data = load_dataset("uoft-cs/cifar10")
```

```python
cifar_train_split = cifar_data["train"]
```

```python
num_images = 5
fig, axes = plt.subplots(1, num_images, figsize=(15, 3))
random_indices = np.random.randint(0, len(cifar_train_split), num_images)
label_names = cifar_train_split.features["label"].names

for i, index in enumerate(random_indices):
    example = cifar_train_split[index]
    image = example["img"]
    label_index = example["label"]
    label_name = label_names[label_index]
    axes[i].imshow(image)
    axes[i].set_title(label_name, fontsize=12)
    axes[i].axis("off")

plt.tight_layout()
plt.savefig('images/cifar10example.png')
plt.show()
```
